**PYTHON OPERATORS:**

```python
#ARITHMETIC OPERATORS: (+,-,*,/,%,**)
X = 5
Y = 10
print(X+Y)
print(X-Y)
print(X*Y)
print(X/Y)
print(X%Y)
print(X**Y)
```

```python
# ASSIGNMENT OPERATORS: (=,+=,-=,*=,/=,%=,**=)
# USED TO ASSIGN TO VALUES:
x,y = 5,8

x+=y # x=x+y
x-=y # x=x-y
x*=y # x=x*y
x/=y # x=x/y
x%=y # x=x%y
x**=y # x=x**y
```

```python
print(x)
```

```python
#Initial: x = 5, y = 8
x += y → x = 5 + 8 = 13
x -= y → x = 13 − 8 = 5
x *= y → x = 5 × 8 = 40
x /= y → x = 40 ÷ 8 = 5.0
x %= y → x = 5.0 % 8 = 5.0
x **= y → x = 5.0 ** 8 = 390625.0

Final value of x = 390625.0
```
