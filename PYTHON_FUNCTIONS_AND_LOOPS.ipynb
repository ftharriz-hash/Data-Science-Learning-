PYTHON FUNCTIONS:


*   A function is an block of code which is only runned when it is called.
*   Advanced functions include lambda,map and zip.



```python
#FUNCTION DEFINING:

def my_function():
  print ("Hello i'm in function")
```

```python
#FUNCTION CALLING:
my_function()
```

```python
#FUNCTION WITH ARGUMENTS:

def my_ID(name):
  print("0019"+ name)

  my_ID("vizam")
```

```python
my_ID()
```

```python
def add(a,b,c):
  return a+b+c

add(9,9,9)
```

```python
#USING LAMBDA FUNCTION:

add = lambda a,b,c: print(a-b+c)
add(9,9,7)
```

```python
#MAP FUNCTION:
nums = [1,2,3,4,5]

def sq(n):
  return n*n

  square = map(sq,nums)
  print(square)
```

```python
#ZIP FUNCTION:
name = ["dilip","nisha","raj"]
ID = [4,2,1]

mapped = zip(name,ID)
print(list(mapped))

```

PYTHON LOOP FUNCTIONS (FOR LOOP):

*   It is used for iterating over a sequence (that is either a list, a tuple, a dict , a set or a string).
*   Using a break and continoue keyword.



```python
#THE ITERATION:
fruits = ["apple","banana","cherry","pineapple"]

for x in fruits:  # x in the place of iteration.
 print(x)
```

```python
#THE BREAK FUNCTION:
for x in fruits:
  print(x)
  if x == "banana":
    break
```

```python
#THE CONTINOUE FUNCTION:
for x in fruits:
  print(x)
  if x == "banana":
    continue
```

WHILE LOOP:


*   With the while loop we can execute a set of statements as long as a condition is true.
*   Using a break and continue functions.



```python
num = 1
while num == 1:
  print(num)
  num += 1
```

```python
#USING BREAK FUNCTION:
num = 1
while num < 11:
  print(num)
  if num == 5:
    break
  num += 1
```

```python
#USING CONTINOUE FUNCTION:
num = 1
while num < 11:
  num += 1
  if num != 5:
    continue
  print(num)
```
