**List and it's Methods:

Lists are used to store multiple items in a single variable.
- Lists are used to store multiple items in a single variable.
- It is a collection which is ordered, changeable and allows duplicate members.

List Methods:
append(), sort(), reverse(), remove(), pop(), insert(), index(), extend(), count(), copy(), clear()**




```python
# APPEND: (A FUNCTION WHICH IS USED TO ADD AN DATA)

fruits = ['apple','cherry','banana']

fruits.append('orange')
print(fruits)
```

```python
# SORT (A FUNCTION IS USED TO SORT THE LIST IN ASCENDING ORDER)

fruits.sort()
print(fruits)
```

```python
# REVERSE (IT REVERSES THE LIST IN VARIABLES):

fruits.reverse()
print(fruits)
```

```python
# REMOVE (THIS FUNCTION REMOVES THE ANY DATA AT VARIABLE):

fruits.remove('cherry')
print(fruits)
```

```python
# POP (THIS FUNCTION IS USED TO DELETE THE SPECIFIC ELEMENT):

fruits.pop(2)
print(fruits)
```

```python
# INSERT (ADDS AN ITEM AT AN SPECIFIC POSITION)

fruits.insert(2,'muskmelon')
print(fruits)
```

```python
# INDEX (RETURNS INDEX OF AN ELEMENT)

fruits.index('muskmelon')
print(fruits)
```

```python
#EXPEND (ADDS ELEMENT FROM ANOTHER LIST)

others = ['pineapple','guva','strawberry','cherry','cherry']

fruits.extend(others)
print(fruits)
```

```python
# COUNT (RETURNS THE NUMBER OF ELEMENTS)

fruits.count("cherry")
```

```python
#CLEAR (REMOVES ALL THE ITEMS FROM THE LIST)

fruits.clear("cherry")
print(fruits)
```

```python
#COPY (RETURNS THE COPY OF THE LIST)

fruits.copy()
print(fruits)
```

ðŸ”¹ What is a String?

All string methods return new values â€” they do NOT change the original string (strings are immutable).**

```python
# CAPITALIZE;

text = "hellow welcome to our course"
text = text.capitalize()
print(text)
```

```python
# COUNT

text = "hellow welcome to our course"
text = text.count("hellow")
print(text)
```

```python
# FIND:

text = "hellow welcome to our course"
text = text.find("welcome")
print(text)

# THE "W" IS IN 7 TH PLACE:
```

```python
# INDEX:

text = "hellow welcome to our course"
text = text.index("our")
print(text)
```

```python
# SPLIT:

text = "hellow welcome to our course , course starts on jan 11"
text = text.split(',')
print(text)
```

```python
# ALPHA: (alpha)

text1 = "company_x"
text1 = text1.isalpha()
print(text1)
```

```python
# ALPHANUMERIC:

text2= "company19"
text2 = text2.isalnum()
print(text2)
```

```python
# DIGIT:

text3 = "3336"
text3 = text3.isdigit()
print(text3)
```

```python
# JOIN FUNCTION:

text4 = ["james","bond","007"]
text4 = "_".join(text4)
print(text4)
```

```python
# REPLACE :

text5 = "james bond 008"
text5 = text5.replace("008","007")
print(text5)
```

```python
#ENDSWITH:

txt = "Hello, welcome to Python"
result = txt.endswith("Python")
print(result)
```

```python
names = ["Ravi Kumar", "James Bond", "Peter","magesh Kumar"]
for name in names:
    if name.endswith("Kumar"):
        print(name)
```

PYTHON CONDITIONAL STATEMENTS (if , else , elif)

```python
# EXAMPLE 1:

marks = 85

if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 50:
    print("Grade: C")
else:
    print("Fail")
```

```python
#EXAMPLE 2:

role = "admin"

if role == "admin":
    print("Full Access Granted")
elif role == "manager":
    print("Limited Access Granted")
else:
    print("Access Denied")
```

```python
#EXAMPLE 3:

temp = 32

if temp > 40:
    print("Too Hot!")
elif temp > 20:
    print("Weather is Normal")
elif temp > 10:
    print("It's Cold")
else:
    print("Freezing!")
```
